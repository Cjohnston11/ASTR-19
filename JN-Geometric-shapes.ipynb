```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import patches
```

```python
#create a figure
f = plt.figure(figsize=(6,6))

#add a single subplot to the figure
#with empty x and y tick liables
ax = f.add_subplot(111,xticks=[], yticks=[])

#set the limits
ax.set_xlim([0,1])
ax.set_ylim([0,1])

#blue rectangle 
rectangle = patches.Rectangle((0.05, 0.05), 0.4,0.4,edgecolor="None", facecolor="blue", alpha=0.5)
ax.add_patch(rectangle)

#red circle
circle = patches.Circle((0.6, 0.6), 0.3,edgecolor="None", facecolor="red", alpha=0.5)
ax.add_patch(circle)

#green ellipse
ellipse = patches.Ellipse((0.8, 0.8), 0.3,0.4,edgecolor="None", facecolor="green", alpha=0.5)
ax.add_patch(ellipse)
```

```python
import matplotlib.cm as cm
import numpy as np
```

```python
#create figure
f = plt.figure(figsize=(5,10))

#get turbo color map
cmap_turbo = cm.get_cmap('turbo')
#add a single subplot to the figure, empty x y tickables and no axis shown
ax = f.add_subplot(111,xticks=[], yticks=[])
ax.set_axis_off()

#set xy limits
ax.set_xlim([0, 2*np.pi])
ax.set_xlim([0, 15.])

#create an x arrange 
x = np.linspace(0,2*np.pi,100)

#number of mountain ranges
nmr = 10
#number of peaks in each range
npeak = 10

#loop over mountain ranges 
for i in range(nmr):
    #set color by probiding a number [0,1]
    mr_color = cmap_turbo(float(i)/float(npeak))

    #create random set of frequences
    freq = 3.0*np.random.rand(npeak)

    #create a randome set of phase shifts
    phi = 2.0*np.pi*np.random.rand(npeak)

    #random amplitude
    amp = 1.*np.random.rand(npeak)+0.1

    #y offset so eacah range is offest 
    yoff = 1.*float(i)

    #create mountain range
    y = np.zeros_like(x)

    #for every peak
    for j in range (npeak):
        #addd peak as sin curve with random amplitude frequency and phse
        y += amp[j]*np.sin(freq[j]*x + phi[j])

    #take abs v of sin cruve and add offset
    y = np.fabs(y) + yoff

    #plot random mountain range
    plt.fill_between(x,y,yoff,zorder = nmr-i,color= mr_color,cmap='turbo')

#save the figure
plt.savefig('mountain_range.png',bbox_inches='tight',pad_inches=0,dpi=600)

    
```
