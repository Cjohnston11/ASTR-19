```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

```python
fig = plt.figure(figsize=(7,7)) #7x7 image
ax = fig.add_subplot(111)
x = np.linspace(0,1*np.pi,200)
y = np.linspace(0,2*np.pi,200)
x,y = np.meshgrid(x,y) #all combinations of x and y
z = -np.sin(x)*np.sin(y)
ax = plt.imshow(z,cmap=cm.turbo,origin = 'lower')
cbar = fig.colorbar(ax)
cbar.ax.get_yaxis().labelpad=15
cbar.ax.set_ylabel('Z Value',rotation=270)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.savefig("impage-plot.png", bbox_inches="tight",facecolor="white")
```
